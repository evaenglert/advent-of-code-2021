```python
def make_set(digit):
    set_digit = set()
    for char in digit:
        set_digit.add(char)
    return set_digit

def create_dict(example):
    dictionary = {}
    five = []
    six = []
    for digit in example:
        if len(digit) == 2:
            dictionary[digit] = 1
        elif len(digit) == 3:
            dictionary[digit] = 7
        elif len(digit) == 4:
            dictionary[digit] = 4
        elif len(digit) == 7:
            dictionary[digit] = 8
        elif len(digit) == 5:
            five.append(digit)
        elif len(digit) == 6:
            six.append(digit)

    five_dictionary = {}

    for digit in five:
        for char in digit:
            if char not in five_dictionary.keys():
                five_dictionary[char] = 1
            else:
                five_dictionary[char] += 1

    five_dictionary
    digit_for_three = ""
    for key, value in five_dictionary.items():
        if value != 1:
            digit_for_three += key

    dictionary[digit_for_three] = 3

    for digit in five:
        if make_set(digit) == make_set(digit_for_three):
            five.remove(digit)

    for digit in six:
        if len(make_set(digit) - make_set(digit_for_three)) == 1:
            dictionary[digit] = 9
            six.remove(digit)

    for digit5 in five:
        for digit6 in six:
            if len(make_set(digit6) - make_set(digit5)) == 1:
                dictionary[digit5] = 5
                dictionary[digit6] = 6
                six.remove(digit6)
                five.remove(digit5)

    dictionary[five[0]] = 2
    dictionary[six[0]] = 0

    return dictionary

def find_integer(dictionary, encoded_input):
    solution = ""
    for digit in encoded_input: 
        for key, value in dictionary.items():
            if make_set(digit) == make_set(key):
                solution += str(value)
    return int(solution)
```

```python
all_integers = []

for row in data.split("\n"):
    all_digits = row.split(" | ")[0].split(" ")
    dictionary = create_dict(all_digits)
    encoded_input = row.split(" | ")[1].split(" ")
    integer = find_integer(dictionary, encoded_input)
    all_integers.append(integer)
```
